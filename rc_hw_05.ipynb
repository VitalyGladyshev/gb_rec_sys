## Гибридные рекомендательные системы

- В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?
- Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы
- Прочитайте статью про поиск на hh.ru https://habr.com/ru/company/hh/blog/347276/

__Какие основные отличия предложенной системы от тех подходов, которые мы разбирали на семинарах? Какие проблемы могут возникнуть при выводе такой модели в продакшен?__

### Ответ

Методы коллаборативной фильтрации предполагают, что интересы пользователей представлены оценками, которые они дают объектам после просмотра, покупки и т.д. Основная идея данных методов заключается в сравнении между собой интересов различных пользователей или объектов на основе этих оценок. При этом никакой дополнительной информации о самих пользователях и объектах не используется.

__Коллаборативная фильтрация (Collaborative filtering)__ – это метод рекомендации, при котором анализируется только реакция пользователей на объекты: оценки, которые выставляют пользователи объектам. Оценки могут быть как явными (пользователь явно указывает, на сколько «звездочек» он оценивает объект), так и неявными (например, количество просмотров одного
ролика). Чем больше оценок собирается, тем точнее получатся рекомендации. Получается, что пользователи помогают друг другу в фильтрации объектов. Поэтому такой метод называется также совместной фильтрацией.

В зависимости от того, какие данные используются
для расчета рекомендаций, системы делятся на три больших класса:
- Методы коллаборативной фильтрации
- Методы, анализирующие содержимое объектов
- Методы, основанные на знаниях

Крупные системы рекомендаций (СР), которые используют различные данные об объектах, пользователях, связях между ними, представляют собой сложную __гибридную систему__. __Она является объединением различных подходов и алгоритмов: коллаборативной фильтрации, анализа содержимого, анализа социальных связей и связей между объектами, и т.д.__

__Пример: видеохостинги, интернет-магазины книг типа LitRes, Ozon__

Выделяется три типа гибридных систем:
- С монолитной организацией
- С параллельной организацией
- С конвейерной организацией

__1. Монолитная организация__

В отличие от параллельной и конвейерной организаций гибридных систем, системы с монолитной организацией представляют собой один модуль, включающий в себя различные методы обработки имеющихся данных.

Большинство гибридных систем такого рода реализуют коллаборативную фильтрацию, усиленную особенностями содержимого. Например, если алгоритм коллаборативной фильтрации рекомендует некоторому пользователю несколько книг, то среди них можно выбрать наиболее подходящую с помощью сравнения жанров этих книг с популярными жанрами данного пользователя. Т.е. если пользователь предпочитает фантастику (что легко узнать, просмотрев жанры всех оцененных/прочитанных пользователем книг или его профиль), то среди книг двух жанров: история и фантастика, система предложит второй вариант.

__2. Параллельная организация__

Исходные данные поступают на каждый модуль системы. После того, как все подсистемы выдадут ответ, начинает работать агрегирующий модуль.
<br>Есть три основных подхода выбрать множество наиболее релевантных
рекомендаций:
<br>
- Первый подход – смешать все рекомендации выдать их пользователю;
- Второй подход – для каждого объекта подсчитать сумму рейтингов, полученных параллельными модулями и вывести объекты с наибольшим рейтингом;
- Третий подход – выбрать одну подходящую систему рекомендации, реализованную в одном из модулей для каждой конкретной ситуации. Например, если модуль коллаборативной фильтрации не дает ответ из-за нехватки данных (проблема «холодного старта»), система выбирает метод анализа содержимого.

__3. Конвейерная организация__

Задача рекомендации делится на этапы, которые выполняются последовательно. Результат работы очередного этапа используется в последующих.

__Пример: система рекомендаций сайта hh.__ Статья: "Умный поиск: как искусственный интеллект hh.ru подбирает вакансии к резюме".

## LightFM

__На вебинаре мы рассматривали модель LightFM (https://making.lyst.com/lightfm/docs/lightfm.html). В работе Data Scientist'а важную часть занимает research - исследование существующих архитектур и разбор научных статей, в которых они описываются. Вам предлагается изчуть оригинальную статью про LightFM https://arxiv.org/pdf/1507.08439.pdf и ответить на следующие вопросы:__

1. Какой датасет используют авторы?
2. Что используют в качестве признаков?
3. С какими моделями сравнивают LightFM? Опишите их основные идеи кратко

### Ответ

Для демонстрации алгоритма используются два датасета:

- MovieLens
- CrossValidated

__MovieLens__
<br>В первом эксперименте используется хорошо известный набор данных MovieLens 10M в сочетании с набором тегов Tag Genome. Набор данных состоит из примерно 10 миллионов оценок фильмов, присланных 71567 пользователями 10681 фильму. Все фильмы описываются по жанрам и списком тегов из Tag Genome. Каждая пара тегов фильма сопровождается показателем релевантности (от 0 до 1), обозначающим, насколько точно данный тег описывает фильм.

Tag Genome делают двоичным, считают все оценки ниже 4.0 (по шкале от 1 до 5) отрицательными; все оценки, равные или выше 4.0, положительными. Отсеивают все рейтинги, которые падают ниже порога релевантности 0.8, чтобы оставить только самые релевантные теги. Подготовленный набор данных содержит 69878 пользователей, 10681 item, 9996948 взаимодействий и 1030 уникальных тегов.

__CrossValidated__
<br>Второй набор данных состоит из вопросов и ответов, размещенных на CrossValidated, части более крупной сети совместных сайтов вопросов и ответов StackExchange, ориентированных на статистику и машинное обучение. Набор данных состоит из 5953 пользователей, 44200 вопросов и 188865 ответов и комментариев. Каждый вопрос сопровождается одним или несколькими из 1032 уникальных тегов (например, «регрессия» или «проверка гипотез»). Кроме того, метаданные пользователя доступны в виде разделов «Обо мне» в профилях пользователей. 

Цель рекомендации - подобрать пользователям вопросы, на которые они могут ответить. Пользователь, отвечающий на вопрос, воспринимается как неявный положительный сигнал; все вопросы, на которые пользователь не ответил, рассматриваются как неявные негативные сигналы. Для обучающих и тестовых наборов создают 3 отрицательные обучающие пары для каждой положительной пары пользователь-вопрос путем случайной выборки из всех вопросов, на которые данный пользователь не ответил.

С каждым датасетом проводятся два эксперимента: имитация warm-start (20% всех пар взаимодействий случайным образом назначаются тестовой выборке, но все элементы и пользователи представлены в тренировочной выборке) и имитация cold-start (все взаимодействия, относящиеся к 20% item'ов, удаляются из обучающего набора и добавляются в тестовый набор)

__Автор проводит сравнение модели LightFM с моделями основанными на содержимом-контенте (Content-based (CB)) и с моделями, основанными  на матричной факторизации (Matrix factorisation (MF))__

Взаимосвязь между Light FM и моделью collaborative MF определяется структурой наборов фичей пользователей и item'ов. Если фичи состоят исключительно из индикаторных переменных для каждого пользователя и item'а, LightFM сводится к стандартной модели MF. Если фичи также содержат метаданные, общие для нескольких item'ов или пользователей, LightFM расширяет модель MF, позволяя фичам скрытых факторов объяснять часть структуры взаимодействий пользователей.

Когда присутствуют только фичи основанные на метаданных и нет индикаторных переменных, модель в целом не сводится к чистой модели основанной на контенте CB. LightFM оценивает эмбединги фичей путем факторизации матрицы совместного взаимодействия; это не похоже на системы, основанные на контенте, которые (когда используется уменьшение размерности) факторизируют чистые контентные матрицы совстречаемости.

Тот факт, что LightFM содержит как чистую контентную модель CB на разреженном данных, так и модель матричной факторизации MF на насыщенных данных, говорит о том, что она должна хорошо адаптироваться к наборам данных различной разреженности. На самом деле, практика показывает, что LightFM работает так же хорошо, как и соответствующая специализированная модель в каждом сценарии.

__Приводится оценка следующих моделей на данных имитирующих warm-start и cold-start:__

1. __MF__ модель обычной матричной факторизации со сдвигами (bias) пользователей и предметов и с сигмоидной функцией.

2. __LSI-LR__ модель, основанная на содержании-контенте. Извлекаются скрытые темы из матрицы item-фичей посредством латентной семантической индексации и представляются item'ы как линейные комбинации латентных тем. Затем создаютя отдельные модели логистической регрессии (LR) для каждого пользователя. В отличие от модели Light FM, которая использует колаборативные данные для создания своего латентного представления, LSI-LR основана исключительно на факторизации матрицы контента.

3. __LSI-UP__ гибридная модель, представляющая пользовательские профили (UP) в виде линейных комбинаций контентных векторов item'ов, с последующим применением LSI к результирующей матрице для получения латентных представлений user'ов и item'ов.

4. __LightFM (tags)__ модель LightFM только с фичами основанными на тегах.

5. __LightFM (tags + ids)__ модель LightFM с фичами на тегах и индикаторных item фичах (ids) 

6. __LightFM (tags + about)__ модель Light FM с использованием как item, так и user фичей. Фичи пользователей можно получить только в датасете CrossValidated из раздела "About Me"
